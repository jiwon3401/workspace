### 04_Apply,Map

```python
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv")
df.drop('Unnamed: 0', axis=1, inplace=True)
df
```

```python
df['Income_Category'].unique()
```

```python
#Income_Category의 카테고리를 map 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라
new_dict = {'Unknown':'N',
       'Less than $40K': 'a',
       '$40K - $60K': 'b',
       '$60K - $80K': 'c',
       '$80K - $120K':'d',
       '$120K +': 'e'}
    
df['newIncome'] = df['Income_Category'].map(lambda x: new_dict.get(x,0))
```

```python
#Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라
def change_category(columns):
    if columns == 'Unknown': return 'N'
    elif columns == 'Less than $40K':return 'a'
    elif columns == '$40K - $60K': return 'b'
    elif columns == '$60K - $80K': return 'c'
    elif columns == '$80K - $120K': return 'd'
    elif columns == '$120K +': return 'e'
    else: print('Error')

df['newIncome']= df['Income_Category'].apply(change_category)
```

```python
df['newIncome'].unique()
```

```python
df['Customer_Age'].unique()
```

```python
df.loc[df['Customer_Age']<30].shape
```

```python
#Customer_Age의 값을 이용하여 나이 구간을 AgeState 컬럼으로 정의하라. (0~9 : 0 , 10~19 :10 , 20~29 :20 … 각 구간의 빈도수를 출력하라

df['AgeState']= pd.cut(df['Customer_Age'], bins = range(10,81,10), labels=range(10,80,10), right=False)
df['AgeState'].value_counts().sort_index()

#df['AgeState'] = df.Customer_Age.map(lambda x: x//10 *10)
```

```python
#Education_Level의 값중 Graduate단어가 포함되는 값은 1 그렇지 않은 경우에는 0으로 변경하여 newEduLevel 컬럼을 정의하고 빈도수를 출력하라
def edu_column(x):
    return 1 if 'Graduate' in x else 0

df['newEduLevel'] = df['Education_Level'].apply(edu_column)
df['newEduLevel'].value_counts()
#df['Education_Level'].map(lambda x : 1 if 'Graudate' in x else 0)
```

```python
#Credit_Limit 컬럼값이 4500 이상인 경우 1 그외의 경우에는 모두 0으로 하는 newLimit 정의하라. newLimit 각 값들의 빈도수를 출력하라

df['newLimit'] = df['Credit_Limit'].map(lambda x: 1 if x>=4500 else 0)
df['newLimit'].value_counts()
```

```python
def condition_c(df):
    if df['Marital_Status'] == 'Married' and df['Card_Category']=='Platinum': return 1
    else: return 0

df['newState'] = df.apply(condition_c, axis=1)
df['newState'].value_counts()
```
